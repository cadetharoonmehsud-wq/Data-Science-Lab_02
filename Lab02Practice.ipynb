## Self Code for mean

```python
totalsum=sum(Dataset_Marks)
print(totalsum)
totallength=len(Dataset_Marks)
print(totallength)
```

```python
# mean
mean=totalsum/totallength
print(mean)
```

## SELF CODE FOR VARIANCE

```python
sumsqdiff=0
for mark in Dataset_Marks:
    diff=mark-mean
    squared=diff**2
    sumsqdiff +=squared
variance=sumsqdiff/(totallength-1)
print(variance)
```
