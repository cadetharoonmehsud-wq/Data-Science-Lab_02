```python
import pandas as pd
```

```python
Dataset_Marks=[23,4,75,87,98,9,65,54,43,65,76,87,34,23,64,76,87,98,67]
df=pd.DataFrame(Dataset_Marks,columns=['Marks'])
print("Marks: ", Dataset_Marks)
```

```python
print(df)
```

```python
bernoulli_data=[1 if mark>=35 else 0 for mark in Dataset_Marks]
print(bernoulli_data)
```

```python
Success=sum(bernoulli_data)
print("Passed Students: ", Success)
Observations=len(bernoulli_data)
print("Total Students: ", Observations)
```

```python
p=Success/Observations
print("Mean:",p)
```

```python
q=1-p
variance=p*q
print("Variance:",variance)
```
